# IPL-2026 DATA ANALYSIS

```python
import pandas as pd              # for data loading and manipulation
import plotly.express as px      # for creating interactive charts easily
import plotly.graph_objects as go # for creating custom/advanced charts
```

```python
# importing dataset file and save in data name variable..
data=pd.read_csv("1-39_Record_Dataset-IPL2026.csv")
data
```

# NUMBER OF MATCHES WON BY EACH TEAM IN IPL 2026

```python
figure = px.bar(
    data,
    x="match_winner",  # Team names on the X-axis
    title="Number of Matches Won by Each Team",  # Chart title
    labels={
        "match_winner": "Match Winner",  # Label for X-axis
        "count": "Matches Won"           # Label for Y-axis
    }
)

figure.show()  # Display the bar chart
```

```python
# Count the number of matches won by defending a target (won by runs)
wins_by_runs = data['wb_runs'].notna().sum()

# Count the number of matches won while chasing a target (won by wickets)
wins_by_wickets = data['wb_wickets'].notna().sum()

# Create a donut chart to show win distribution
fig = go.Figure(
    data=[
        go.Pie(
            labels=['Won by Defending', 'Won by Chasing'],  # Category labels
            values=[wins_by_runs, wins_by_wickets],         # Corresponding counts
            hole=0.4,                                       # Creates a donut chart
            textinfo='label+percent+value'                  # Show label, percentage, and count
            )
        ]
)

# Add chart title
fig.update_layout(
    title='Matches Won by Defending vs Chasing'
)

# Display the chart
fig.show()
```

```python
figure=px.bar(
                data,
                x="best_bowling",  # X-axis: bowler names from dataset
                title="Best Bowler Awards by Bowler - IPL 2026",  # chart title
                 labels={
                        "best_bowling": "Bowler Name",  # rename x-axis label
                        "count": "Best Bowler Awards"   # rename y-axis label
                        }
            )
figure.show()  # display the chart
```

```python
figure=px.bar(
                data,
                x="player_of_the_match",  # X-axis: player names from dataset
                title="Player of the Match Awards - IPL 2026",  # chart title
                labels={
                    "count":"Player of the Match Awards",  # rename y-axis label
                    "player_of_the_match":"Player Name"    # rename x-axis label
                }
            )
figure.show()  # display the chart
```

```python
figure=px.bar(
                data,
                x="top_scorer",    # X-axis: player names from dataset
                y="highscore",     # Y-axis: highest score of that player
                title="Most Top Scorer Awards - IPL 2026",  # chart title
                color="highscore", # bar color changes based on highscore value
                labels=
                    {
                        "top_scorer":"Player Name",  # rename x-axis label
                        "highscore":"HighScore"      # rename y-axis label
                    }
            )
figure.show()  # display the chart
```
